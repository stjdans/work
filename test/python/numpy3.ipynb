```python
import numpy as np
import matplotlib.pyplot as plt

A = np.random.standard_normal(200_000)
print('평균 : ', np.mean(A))
print('표준 편차', np.std(A))
A = A * 10 + 100
print('평균 : {}, 표준 편차 : {}'.format(np.mean(A), np.std(A)))

plt.hist(A, bins=80, color='g')
plt.title('Normal Distruibution in 80 bin')
plt.show()

B = np.histogram(A, 50)
plt.plot(B[0])
# plt.plot(B[1])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.random.standard_normal(200_000)
A = A * 10 + 100
B, X = np.histogram(A, 50)

print()
print(X)
print((X[1:]+X[:-1]))
print('B len : {} X len: {}'.format(len(B), len(X)))
X = (X[1:]+X[:-1]) / 2
plt.plot(X, B)
plt.show()
```
