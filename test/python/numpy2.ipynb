```python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(0, 2 * np.pi)
print(A)

print()

plt.plot(A, np.sin(A))
plt.show()

plt.plot(A, np.cos(A))
plt.show()

plt.plot(A, np.tan(A))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(0, 2 * np.pi)
# plt.plot(A, np.sin(A))
# plt.plot(A, np.cos(A))
# plt.plot(A, np.tan(A))
plt.plot(A, np.sin(A), '*', A, np.cos(A), 'o')
plt.title('Sine and Cosine')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(0, 60, 60)
plt.plot(A, A * 2, 'o')
plt.plot(A, 1.1**A)
plt.title('Compounded Interest v. Linear')
plt.xlabel('Year')
plt.ylabel('Interest')
plt.show()
```
